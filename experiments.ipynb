```python
from commander import Commander
import hydra
from omegaconf import OmegaConf

# Set the function to load and override config.
def load_config(overrides=[]):
    with hydra.initialize(version_base=None, config_path="./config/"):
        return hydra.compose(
            config_name="base.yaml", overrides=overrides
            )

conf = load_config(overrides=[])
print(OmegaConf.to_yaml(conf))

exp = Commander(conf=conf)
exp.run_preprocessor()
exp.run_provider(test_single=True)
```
